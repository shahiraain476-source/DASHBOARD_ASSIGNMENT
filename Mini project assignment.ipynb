Objective
1. To investigate the relationship between study hours, attendance rates, and final exam scores to identify key drivers of student success.
2. To evaluate the impact of lifestyle choices, specifically sleep hours and physical activity levels, on academic outcomes and motivation.
3. To assess the effectiveness of external support systems, such as tutoring sessions and internet access, in improving the exam scores of students with different learning needs.

```python
#1. IMPORT LIBRARY
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#2. LOADING DATA
df = pd.read_csv("StudentPerformanceFactors.csv")
print(df)
```

```python
#3. CREATE A DATA FRAME
data = pd.DataFrame(df)
data
```

```python
#4. RENAMING COLUMN
data = data.rename(columns={'Hours_Studied': 'Hours Studied', 'Parental_Involvement' : 'Parental Involvement', 'Access_to_Resources':'Access to Resources', 'Extracurricular_Activities':'Extracurricular Activities', 'Sleep_Hours': 'Sleep Hours', 'Previous_Scores': 'Previous Scores', 'Motivation_Level': 'Motivation Level', 'Internet_Access' : 'Internet Access', 'Tutoring_Sessions' : 'Tutoring Sessions' , 'Family_Income' : 'Family Income', 'Teacher_Quality': 'Teacher Quality', 'School_Type': 'School Type', 'Peer_Influence': 'Peer Influence', 'Physical_Activity': 'Physical Activity', 'Learning_Disabilities' :'Learning Disabilities',  'Parental_Education_Level': 'Parental Education Level', 'Distance_from_Home' : 'Distance from Home', 'Exam_Score' : 'Exam Score' })
data
```

```python
#5. CHECK DATA TYPE AND MISSING VALUE
data.info
```

```python
#6. CHECK MISSING VALUE
data.isnull().sum()
```

```python
#7. REMOVE MISSING VALUE
data.dropna(inplace = True)
data
```

```python
#8. CHECK HOW MANY DUPLICATE
data.duplicated().sum()
```

```python
#Repair the wrong data
data = data[data['Exam Score'] <= 100]
```

```python
#9. CHECK SUMMARY STATISTIC
data.describe()
```

```python
#10. CALCULATE MINIMUM, MAXIMUM, MEAN AND STANDARD DEVIATION
summary = data[['Hours Studied', 'Attendance', 'Sleep Hours','Previous Scores', 'Exam Score']].agg(['min', 'max', 'mean','std'])
print(summary)
```

```python
#11. CHANGE summary_stats TO INTEGER
summary = summary.astype(int)
data.info()
```

```python
data.to_csv('clean_data.csv', index=False)
```

## OBJECTIVE 1:
### RELATIONSHIP BETWEEN STUDY HOURS, ATTENDANCE AND EXAM SCORE TO IDENTIFY THE KEY OF STUDENTS SUCCESS

```python
#12. CALCULATE CORRELATION (CHECK RELATIONSHIP = USE CORRELATION)
corr1 = data[['Hours Studied','Attendance','Exam Score']].corr()
print(corr1)
```

```python
#13. AVERAGE SCORES BY HOURS STUDIED
hours_score = data.groupby('Hours Studied').agg({'Previous Scores': 'mean', 'Exam Score': 'mean'})
print(hours_score)
```

```python
#14. VISUALIZE HOURS STUDIED VS EXAM SCORE
fig, ax = plt.subplots(figsize = (8, 6))
#Scatter plot
data.plot(kind='scatter', #Make a scatter graph
        x='Hours Studied', #x axis
        y='Exam Score', # y axis
        color = 'purple', # point color purple
        ax=ax) #draw on the axes created earlier
#Graph title
plt.title('Relationship between Study Hours and Exam Score')
plt.show()
```

```python
#15. VISUALIZE ATTENDANCE VS EXAM SCORE
fig, ax = plt.subplots(figsize = (8, 6))
#Scatter plot
data.plot(kind='scatter', #Make a scatter graph
        x='Attendance', #x axis
        y='Exam Score', # y axis
        color = 'pink', # point color pink
        ax=ax) # draw on the axes created earlier
#Graph title
plt.title('Relationship between Attendance and Exam Score')
plt.show()
```

## OBJECTIVE 2:
### EFFECT OF SLEEP HOURS AND PHYSICAL ABILITY ON ACADEMIC PERFORMANCE

```python
#16. CALCULATE CORRELATION (CHECK RELATIONSHIP = USE CORRELATION)
corr2 = data[['Sleep Hours','Physical Activity','Exam Score']].corr()
print(corr2)
```

```python
#17. AVERAGE EXAM SCORE BY SLEEP HOURS
sleep_score = data.groupby('Exam Score')['Sleep Hours'].mean()
print(sleep_score)
```

```python
#18. AVERAGE EXAM SCORE VS PHYSICAL ACTIVITY
activity_score = data.groupby('Exam Score')['Physical Activity'].mean()
print(activity_score)
```

```python
#19. VISUALIZE SLEEP HOURS VS EXAM SCORE
fig, ax = plt.subplots(figsize = (10, 6))
sleep_score.plot(kind='line',
                 x='Sleep Hours',
                 y='Average Exam Score',
                 marker='o', 
                 color='purple',
                 linewidth= 2,
                 figsize=(8,5),
                 ax =ax)
#Graph title
plt.title('Average Sleep Hours by Exam Score')
plt.show()
```

```python
#20. VISUALIZE PHYSICAL ACTIVITY VS EXAM SCORE
fig, ax = plt.subplots(figsize = (10, 6))
activity_score.plot(kind='line',
                 x='Physical Activity',
                 y='Average Exam Score',
                 marker='o', 
                 color='green',
                 linewidth= 2,
                 figsize=(8,5),
                 ax =ax)
#Graph title
plt.title('Average Physical Activity by Exam Score')
plt.show()
```

## OBJECTIVE 3: 
### EFFECT OF TUTORING SESSIONS AND INTERNET ACCESS ON EXAM SCORE

```python
#21. AVERAGE EXAM SCORE BY TUTORING SESSIONS
tutoring_score = data.groupby('Tutoring Sessions')['Exam Score'].mean()
print(tutoring_score)
```

```python
#22. AVERAGE EXAM SCORE BY INTERNET ACCESS
internet_score = data.groupby('Internet Access')['Exam Score'].mean()
print(internet_score)
```

```python
#23. VISUALIZE TUTORING SESSIONS VS EXAM SCORE
fig, ax = plt.subplots(figsize = (10, 6))
tutoring_score.plot(kind='line',
                    x='Tutoring Sessions',
                    y='Exam Score',
                    marker= 'o', 
                    color='blue',
                    linewidth= 2,
                    figsize=(8,5),
                    ax =ax)
plt.title('Average Exam Score by Tutoring Sessions')
plt.show()
```

```python
#24 VISUALIZE INTERNET ACCESS VS EXAM SCORE
fig, ax = plt.subplots(figsize = (10, 6))
internet_score.plot(kind='barh',
                   x='Internet Access',
                   y='Exam Score', 
                   color='blue',
                   edgecolor='black',
                   ax =ax)
plt.title('Average Exam Score by Internet Access')
plt.show()
```

## DASHBOARD CODING

```python
#IMPORT LIBRARY
import streamlit as st
import pandas as pd
import matplotlib.pyplot as plt

# PAGE TITLE
st.title("STUDENTS PERFORMANCE DASHBOARD")
st.write("ANANLYSIS OF FACTORS AFFECTING STUDENT EXAM PERFORMANCE")

#LOADING DATA
data= pd.read_csv("StudentPerformanceFactors.csv")

#DISPLAY RAW DATA
st.subheader("RAW DATA")
st.dataframe(data)

# OBJECTIVE 1
# HOURS STUDIED VS EXAM SCORE
st.subheader("HOURS STUDIED VS EXAM SCORE")
fig, ax = plt.subplots(figsize = (8, 6))
#Scatter plot
data.plot(
        kind='scatter',
        x='Hours Studied', 
        y='Exam Score', 
        color = 'purple',
        ax=ax
)
st.pyplot(fig)

#ATTENDANCE VS EXAM SCORE
st.subheader("ATTENDANCE VS EXAM SCORE")
fig, ax = plt.subplots(figsize = (8, 6))
data.plot(
        kind='scatter',
        x='Attendance', 
        y='Exam Score', 
        color = 'pink',
        ax=ax
)
st.pyplot(fig)

#OBJECTIVE 2
#SLEEP HOURS
st.subheader("AVERAGE SLEEP HOURS BY EXAM SCORE")
sleep_score = data.groupby('Exam Score')['Sleep Hours'].mean()
fig, ax = plt.subplots(figsize = (10, 6))
sleep_score.plot(kind='line',
                 x='Exam Score',
                 y='Average Sleep Hours',
                 marker='o', 
                 color='purple',
                 linewidth= 2,
                 figsize=(8,5),
                 ax =ax)
st.pyplot(fig)

#PHYSICAL ACTIVITY
st.subheader("AVERAGE PHYSICAL ACTIVITY BY EXAM SCORE")
activity_score = data.groupby('Exam Score')['Physical Activity'].mean()
fig, ax = plt.subplots(figsize = (10, 6))
sleep_score.plot(kind='line',
                 x='Exam Score',
                 y='Average Physical Activity',
                 marker='o', 
                 color='magenta',
                 linewidth= 2,
                 figsize=(8,5),
                 ax =ax)
st.pyplot(fig)

#OBJECTIVE 3
#TUTORING SESSIONS
st.subheader("AVERAGE EXAM SCORE BY TUTORING SESSIONS")
tutoring_score = data.groupby('Tutoring Sessions')['Exam Score'].mean()
fig, ax = plt.subplots(figsize = (10, 6))
tutoring_score.plot(kind='line',
                    x='Tutoring Sessions',
                    y='Exam Score',
                    marker= 'o', 
                    color='maroon',
                    linewidth= 2,
                    figsize=(8,5),
                    ax =ax)
st.pyplot(fig)

#INTERNET ACCESS
st.subheader("AVERAGE EXAM SCORE BY INTERNET ACCESS")
internet_score = data.groupby('Internet Access')['Exam Score'].mean()
fig, ax = plt.subplots(figsize = (10, 6))
internet_score.plot(kind='barh',
                   x='Internet Access',
                   y='Exam Score', 
                   color='violet',
                   edgecolor='black',
                   ax =ax)
st.pyplot(fig)
```
